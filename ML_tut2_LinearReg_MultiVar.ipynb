01 - Data Preprocessing

```python
# importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# loading the data
df = pd.read_csv("mldata2.csv")
df
```

To handle the missing values, I will take the median. It depends on the column whether to take the median or mean to fill the missing values.

First, the distribution of that column should be checked using "df['column_name'].hist()". If the distribution is skewed, its always a good idea to use median as this is robust to outliers. Whereas for symmetric distribution, mean is also ok.

Also, Check the %age of missing values:
If very few missing (<5%); median/mean is fine.
If many missing; consider more advanced methods

```python
# df.bedrooms.median()
#Since our data has no decimal values, so to have simple zero decimal value median, we will use floor function.
import math
median_value = math.floor(df.bedrooms.median())
median_value
```

```python
#Now, fill the missing values of the bedroom column with median_value
df.bedrooms.fillna(median_value)
```

```python
#Now to assign this back to original dataframe, we do this.
df.bedrooms = df.bedrooms.fillna(median_value)
df
```

So, till now, our Data Preprocessing step is Over.

02 - Linear Regression using Multi-Variable

```python
# training the model
reg = linear_model.LinearRegression() #created an object for the model
reg.fit(df[['area', 'bedrooms', 'age']],df.price) #remember X_train is always 2D here
```

```python
# predicting the price of the house based on the 3 variables
reg.predict([[2500,3,10]]) #area=2500, bedrooms= 3, age=10
# also remember the argument to predict function is always given as 2D array.
```

```python
#always a good idea to know the coefficients and intercept to cross check
reg.coef_
```

```python
reg.intercept_
```

```python
#crosschecking
#putting into price = theta1*area + theta2*bedrooms + theta3*age
(2500*144.92857143) + (3*-27517.34693878) + (10*(-4776.53061224)) + (442845.9183673457)
```

Analysis:

Notice that, with a house of area=2500, bedrooms= 3,and age=10. The price is higher than what we trained to the model in row 2 as area = 2600, bedrooms = 2, and age = 23; although the area of that row 2 is more. This is because, the age of predicted model is just 10 which is way less than the age of the house sample given in training with age = 20. Therefore, our model has correctly learnt the relation, did the right trade-off, and given the right price value to the house with the age = 10.

Since our data contains missing values, our first step should be Data Preprocessing.

01 - Data Preprocessing:

Notice here that the experience column is a word column although its literal meaning is a number. So, it would be best idea to convert words to number using "word2number" library. For the missing values in the experience column, we will assume 0 experience.

To handle missing values in test_score column, we will take the median

```python
!pip install word2number
```

```python
from word2number import w2n

#First, filling the experience column with string value 'zero'
df2['experience'] = df2['experience'].fillna('zero')
df2['experience']
```

```python
#now converting all the string words to numbers
df2['experience'] = df2['experience'].apply(lambda x: w2n.word_to_num(x))
df2['experience']
```

```python
# now converting filling the missing values of "test_score(out of 10)" column using median
medvalue = df2['test_score(out of 10)'].median()
medvalue
```

```python
df2['test_score(out of 10)'] = df2['test_score(out of 10)'].fillna(medvalue)
df2
```

Till here, Data Preprocessing is done.

02 - Linear Regression using Multi Variables

```python
# training the model
reg2 = linear_model.LinearRegression()
reg2.fit(df2[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']], df2[['salary($)']])
```

```python
# predicting the salary with experience = 12, test_score = 10, interview_score = 10
reg2.predict([[12,10,10]])
```
